```python
# 12-3
#######################################################################
import matplotlib as mpl

mpl.rcParams['axes.unicode_minus'] = False
mpl.rc('font', family = 'Malgun Gothic')

font1 = {'size' :18, 'color': 'green'}
####################################################한글설정###########

xdata = ['안지영','홍지수','황예린']
ydata = [90,85,88]
plt.plot(xdata,ydata)

plt.title('세명 학생의 국어 성적', fontdict=font1)
plt.xlabel('이름')
plt.ylabel('성적')

plt.show()
```

```python
# 12-4
xdata = ['안지영','홍지수','황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]
plt.plot(xdata,ydata1,label = '국어')
plt.plot(xdata,ydata2,label = '영어')

plt.legend(loc = 'upper center')

plt.title('세명 학생의 국어, 영어 성적')

plt.show()
```

```python
# 12-5 Line Style
xdata = ['안지영','홍지수','황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]
ydata3 = [85,97,78]
ydata4 = [92,88,82]
                     #color marker linestyle도 가능=>
                  # ex)- 'ro-'
plt.plot(xdata, ydata1, label='국어', color = 'r', linestyle='-', marker='o')
plt.plot(xdata, ydata2, label='영어', color = 'c', linestyle='--', marker='x')
plt.plot(xdata, ydata3, label='수학', color = 'm', linestyle='-.', marker='s')
plt.plot(xdata, ydata4, label='사회', color = '#444444', linestyle=':', marker='d')

plt.title('세명 학생의 네 과목 성적')
plt.legend(loc='best')
plt.show()
```

```python
xdata = ['안지영','홍지수','황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]
ydata3 = [85,97,78]
ydata4 = [92,88,82]
                     #color marker linestyle도 가능=>
                  # ex)- 'ro-'
plt.plot(xdata, ydata1,'ro-' , label='국어',linewidth = 5, markersize=10)
plt.plot(xdata, ydata2,'cx--',label='영어',linewidth = 8, markersize=10)
plt.plot(xdata, ydata3,'ms-.',label='수학',linewidth = 5, markersize=10)
plt.plot(xdata, ydata4,'md:',label='사회',linewidth = 5, markersize=10)

plt.title('세명 학생의 네 과목 성적')
plt.legend(loc='best')
plt.show()
```

```python
# 12-6 X/Y축의 범위와 눈금 설정
import matplotlib.pyplot as plt

xdata = list(range(30))
ydata = []

for x in xdata:
    y=2*x
    ydata.append(y)
    
plt.plot(xdata, ydata, 'mo-',label='y=2x',linewidth=8)
plt.xlim(0,35)
plt.ylim(0,70)
plt.xticks(list(range(0,35,2)))  #x틱스 y틱스있는 곳에 그리드가 그려짐
plt.yticks(list(range(0,70,5)))

plt.grid(True)
plt.show()
```

```python
# 12-6 X/Y축의 범위와 눈금 설정
import matplotlib.pyplot as plt

plt.title('y=2x')
plt.plot(xdata, ydata, color='magenta', linewidth = 3)
plt.xlim(-1,30)
plt.ylim(-2, 60)
plt.xticks(list(range(0,30,2)))
plt.yticks(list(range(0,60,5)))
 
plt.grid(True)
plt.show()
```
